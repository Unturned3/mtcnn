```python
# facenet_pytorch_c: avoid confusion with system default facenet_pytorch
from facenet_pytorch_c import MTCNN
import torch
from torch.utils.data import DataLoader
from torchvision import datasets
import numpy as np
import pandas as pd
import os

workers = 0 if os.name == 'nt' else 4
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print('Running on device: {}'.format(device))
```

```python
mtcnn = MTCNN(
    image_size=160, margin=0, min_face_size=20,
    thresholds=[0.6, 0.7, 0.7], factor=0.709, post_process=True, keep_all=True,
    device=device
)
```

```python
def collate_fn(x):
    return x[0]

dataset = datasets.ImageFolder('/home/ubuntu/mtcnn/data')
dataset.idx_to_class = {i:c for c, i in dataset.class_to_idx.items()}
loader = DataLoader(dataset, collate_fn=collate_fn, num_workers=workers)
```

```python
import matplotlib.pyplot as plt
import torchvision
```

```python
aligned = []
#names = []
for x, y in loader:
    x_aligned, prob = mtcnn(x, return_prob=True)
    if x_aligned is not None:
        for k in zip(x_aligned, prob):
            img = torchvision.transforms.ToPILImage()(k[0])
            plt.imshow(img)
            plt.show()
            print('Face detected with probability: {:8f}'.format(k[1]))
            aligned.append(img)
            #names.append(dataset.idx_to_class[y])
```
